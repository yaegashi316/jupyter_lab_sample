```python
import numpy as np
```

```python
x=np.array([11,22,33,44])
```

```python
plt.plot(x, x, 'r--')
plt.plot(x, x**2, 's')
plt.plot(x, x**3, 'g^')
plt.legend()
```

```python
plt.plot(x, x, 'r--', label='y=x')
plt.plot(x, x**2, 's', label='y=x**2')
plt.plot(x, x**3, 'g^', label='y=x**3')
plt.legend()                       
```
